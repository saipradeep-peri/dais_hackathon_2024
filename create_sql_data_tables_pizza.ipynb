# Imports and Connecting to in-Memory Sql

```
import sqlite3
import pandas as pd


# sql_database_path = '/Volumes/workspace/default/databricks_hackathon/sql_database.db'
# Create an in-memory SQLite database
conn = sqlite3.connect(':memory:')
```

```
# Define the CSV file path
csv_file_path = '/Volumes/workspace/default/databricks_hackathon/A_year_of_pizza_sales_from_a_pizza_place_872_68.csv'

# Read the CSV file into a pandas DataFrame
df = pd.read_csv(csv_file_path, index_col=0)

# Sample 10% of the rows from the DataFrame
sampled_df = df.sample(frac=0.1, random_state=1)

# Insert the DataFrame into the SQLite table
sampled_df.to_sql(table_name, conn, if_exists='replace', index=False)

# Verify the data has been inserted
result = pd.read_sql(f"SELECT * FROM {table_name}", conn)
print(result)

# Close the connection
# conn.close()
```

```
# Define the file path for the SQLite database
db_file_path = '/Workspace/Shared/databricks_hackathon_2024/pizza_sales_sql.db'

# Create a connection to the filesystem SQLite database
disk_conn = sqlite3.connect(db_file_path)

# Use the backup API to copy the in-memory database to the filesystem
with disk_conn:
    conn.backup(disk_conn)

# Close the connections
conn.close()
disk_conn.close()
```
